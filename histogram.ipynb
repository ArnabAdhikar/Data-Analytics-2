```python
# importing all the necessary libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
from pandas.plotting import andrews_curves
from pandas.plotting import parallel_coordinates
```

```python
# loading the data
data = pd.read_csv("Iris.csv")
print (data.head(10))
```

```python
# calculating all the statistical data
data.describe()
```

```python
# extracting all the attribute type
data.info()
```

```python
# Histogram for sepal length
plt.figure(figsize = (10, 7)) 
x = data["SepalLengthCm"] 
plt.hist(x, bins = 20, color = "green") 
plt.title("Sepal Length in cm") 
plt.xlabel("Sepal_Length_cm") 
plt.ylabel("Count")
```

```python
# Histogram for sepal width
plt.figure(figsize = (10, 7)) 
x = data.SepalWidthCm 
plt.hist(x, bins = 20, color = "green") 
plt.title("Sepal Width in cm") 
plt.xlabel("Sepal_Width_cm") 
plt.ylabel("Count") 
plt.show() 
```

```python
# Histogram for Petal Length
plt.figure(figsize = (10, 7)) 
x = data.PetalLengthCm 
plt.hist(x, bins = 20, color = "green") 
plt.title("Petal Length in cm") 
plt.xlabel("Petal_Length_cm") 
plt.ylabel("Count") 
plt.show()
```

```python
# Histogram for Petal Width
plt.figure(figsize = (10, 7)) 
x = data.PetalWidthCm
plt.hist(x, bins = 20, color = "green") 
plt.title("Petal Width in cm") 
plt.xlabel("Petal_Width_cm") 
plt.ylabel("Count") 
plt.show()
```

```python
# constructing the pairplots
sns.pairplot(data.drop("Id", axis=1), hue="Species", size=3)
plt.show()
```

```python
# hexagonal bin plot
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=data,ratio=10, kind='hex',color='green')
plt.show()
```

```python
# box plot
new_data = data[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]] 
print(new_data.head())
plt.figure(figsize = (10, 7)) 
new_data.boxplot() 
```

```python
data.drop("Id", axis=1).boxplot(by="Species", figsize=(10, 10))
plt.show()
```

```python
# detecting outliers using andrew's curve.
andrews_curves(data.drop("Id", axis=1), "Species",colormap='rainbow')
plt.show()
```

```python
# Parallel cordinates plot each feature on a seperate column.
parallel_coordinates(data.drop("Id", axis=1), "Species",colormap='cool')
plt.show()
```

```python
# Useing Pandas' radviz to plot features on a 2D plane'
from pandas.plotting import radviz
radviz(data.drop("Id", axis=1), "Species",colormap='autumn')
plt.show()
```
